```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data = pd.read_csv('C:\\Users\\DELL\\Desktop\\internship\\task3\\householdtask3.csv')
```

```python
#Scatter plot with year against own ,scatter plot is use to ovserve relationship with variable and uses dots to represent relationship between them.
plt.scatter(data['year'],data['own'])

#adding title to the plot
plt.title("Scatter plot")

#setting the x and y label
plt.xlabel('year')
plt.ylabel('own')
plt.colorbar()
#Showint the result
plt.show()


```

```python
#Line Chart with year and own, it is used to represent relatinship between two day type and yon different axis
plt.plot(data['year'])
plt.plot(data['own'])
#adding title to the plot
plt.title("Line Chart")

#setting the x and y label
plt.xlabel('year')
plt.ylabel('own')

#Showint the result
plt.show()
```

```python
#Bar chart or Bar plot, it is graph that represent the catogary of data  with rectungular bars with length and hight propertional to the value the represent.
plt.bar(data['year'],data['own'])
#adding title to the plot
plt.title("Bar chart")

#setting the x and y label
plt.xlabel('year')
plt.ylabel('own')

#Showint the result
plt.show()
```

```python
#Histogram
plt.hist(data['income'])

#adding title to the plot
plt.title("Histogram")

#Showint the result
plt.show()
```
