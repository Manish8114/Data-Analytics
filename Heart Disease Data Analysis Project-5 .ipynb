```python
#now lets take a look at our age column
#Create a Distribution Plot With normal distribution curve
sns.displot(x= 'age', data = df, bins= 30, kde = True);
#kde= kirnal density
```

```python
sns.displot(df['age'], bins=30, kde=True)

# Add title and labels
plt.title("Age Distribution with Normal Distribution Curve")
plt.xlabel("Age")
plt.ylabel("Density")
```

```python
#58-59 year old people are most in the dataset.
#lets plot another distribution plot for maximum heart rate

sns.displot(x = 'thalach', data=df,  bins=30, kde=True, color= 'chocolate')
# Add title and labels
plt.title("maximum heart rate")
plt.xlabel("Age")
plt.ylabel("Density");
```

```python
#  5. What is the relationship between age and the likelihood of having heart disease?"
# Scatter plot with regression line
sns.lmplot(x='age', y='target', data=df, logistic=True)

# Add title and labels
plt.title("Relationship Between Age and Heart Disease")
plt.xlabel("Age")
plt.ylabel("Likelihood of Heart Disease (0 = No, 1 = Yes)")

plt.show()
```

```python
# Box plot of age vs. heart disease
sns.boxplot(x='target', y='age', data=df)

# Add title and labels
plt.title("Age Distribution by Heart Disease Status")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Age")

plt.show()
```

```python
import numpy as np

# Calculate correlation
correlation = df['age'].corr(df['target'])
print(f"Correlation between Age and Heart Disease: {correlation:.2f}")
```

```python
# How does blood pressure level affect the risk of heart disease?",
# Scatter plot with regression line
sns.lmplot(x='trestbps', y='target', data=df, logistic=True)

# Add title and labels
plt.title("Relationship Between Blood Pressure and Heart Disease")
plt.xlabel("Resting Blood Pressure (trestbps)")
plt.ylabel("Likelihood of Heart Disease (0 = No, 1 = Yes)")
```

```python
# Box plot of blood pressure vs. heart disease
sns.boxplot(x='target', y='trestbps', data=df)

# Add title and labels
plt.title("Blood Pressure Distribution by Heart Disease Status")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Resting Blood Pressure (trestbps)")
```

```python
# Violin plot of blood pressure vs. heart disease
sns.violinplot(x='target', y='trestbps', data=df)

# Add title and labels
plt.title("Blood Pressure Distribution by Heart Disease Status")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Resting Blood Pressure (trestbps)")
```

```python
# Distribution plot of blood pressure by heart disease status
sns.displot(df, x='trestbps', hue='target', kind='kde', fill=True)

# Add title and labels
plt.title("Blood Pressure Distribution by Heart Disease Status")
plt.xlabel("Resting Blood Pressure (trestbps)")
plt.ylabel("Density")
```

```python
# Calculate correlation
correlation = df['trestbps'].corr(df['target'])
print(f"Correlation between Blood Pressure and Heart Disease: {correlation:.2f}")
```

```python
#  7. What is the impact of cholesterol levels on the prevalence of heart disease?"]

# Distribution plot of cholesterol levels by heart disease status
sns.displot(df, x='chol', hue='target', kind='kde', fill=True)

# Add title and labels
plt.title("Cholesterol Level Distribution by Heart Disease Status")
plt.xlabel("Cholesterol Level (chol)")
plt.ylabel("Density")
```
